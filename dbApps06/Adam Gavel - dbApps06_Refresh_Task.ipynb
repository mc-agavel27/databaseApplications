**Reminder — Key SQL syntax:**
- `SELECT ... FROM ... WHERE ...` — pick columns, filter rows
- `LIKE '%keyword%'` — partial text matching (`%` = any characters)
- `ORDER BY column DESC` — sort results (DESC = largest first)
- `LIMIT 10` — only return 10 rows
- `GROUP BY column` — group rows for `COUNT()`, `AVG()`, etc.
- `JOIN table2 ON table1.col = table2.col` — combine two tables
- `IN ('val1', 'val2')` — match any value in the list
- `IS NOT NULL` — filter out blank values
- `ROUND(value, 2)` — round to 2 decimal places
- `COUNT(DISTINCT col)` — count unique values only

*Remember:* You need to JOIN `name_basics` → `title_principals` → `title_basics` → `title_ratings`

**My actor/actress:** _______________

```python
# Task 4a: How many movies is your actor in?
query = '''
    SELECT n.primaryName, COUNT(DISTINCT b.tconst) as movieCount
    FROM name_basics n
    JOIN title_principals p ON n.nconst = p.nconst
    JOIN title_basics b ON p.tconst = b.tconst
    WHERE n.primaryName = '___'
      AND p.category IN ('actor', 'actress')
      AND b.titleType = 'movie'
    GROUP BY n.nconst
'''

df = pd.read_sql_query(query, connection)
print(df.to_string(index=False))
```

*Hint:* Use `WHERE genres LIKE '%Action%'` to match a genre. Remember genres are stored like `"Action,Adventure,Sci-Fi"` so LIKE is the right tool here.

```python
# Task 5a: Genre 1 stats
query = '''
    SELECT '___' as genre,
           COUNT(*) as movieCount,
           ROUND(AVG(r.averageRating), 2) as avgRating,
           ROUND(AVG(r.numVotes), 0) as avgVotes
    FROM title_basics b
    JOIN title_ratings r ON b.tconst = r.tconst
    WHERE b.titleType = 'movie'
      AND b.genres LIKE '%___%'
'''

df = pd.read_sql_query(query, connection)
print(df.to_string(index=False))
```
